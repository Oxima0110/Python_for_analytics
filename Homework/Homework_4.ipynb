## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
```

```python
corr_matrix = df.corr()
plt.figure(figsize = (20,18))
sns.set(font_scale = 1.4)
sns.heatmap(corr_matrix, annot = True, linewidth = .5, cmap = 'coolwarm')
plt.title('Матрица корреляций')
```

Построив матрицу корреляций мы видим что на цену влияют след. характеристики:
1. Количество комнат
2. Состояние дома
3. Площадь дома
4. Этажность дома
5. Жилая площадь

```python
plt.figure(figsize = (12, 8))
#plt.scatter(df['price'], df['sqft_living'])
sns.scatterplot(x = df['price'], y = df['sqft_living'])
plt.xlabel('Цена')
plt.ylabel('Площадь')
plt.title('Взаимосвязь цены и площади домов');
```

При увеличении площади дома цена увеличивается

```python
plt.figure(figsize=(18, 9))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)

plt.xlabel('Состояние')
plt.ylabel('Цена')
plt.title('Взаимосвязь цены и состояния домов')
plt.show()
```

Чем лучше состояние дома, тем выше его цена

```python
sns.jointplot(x=df['price'], y = df['sqft_above'], kind = 'reg')

plt.title('Взаимосвязь цены и жилой площади домов');
```

С увеличением жилой площади растет и стоимость дома

```python
sns.jointplot(x=df['price'], y = df['bedrooms'], kind = 'kde')
plt.title('Взаимосвязь цены и количества комнат домов');
```

Дома с 2 этажами стоят дороже домов с 1 этажом

```python
plt.figure(figsize=(18, 9))

sns.boxplot(x=df['floors'], y=df['price'], whis=1.5)

plt.xlabel('Этажи')
plt.ylabel('Цена')
plt.title('Взаимосвязь цены и этажности домов')
plt.show()
```

```python
Дома с 2,5 этажами стоят дороже
```

```python
plt.xlabel('Цена')
plt.ylabel('Жилая площадь после ремонта')
plt.scatter(x=df['price'], y=df['sqft_living15'])
plt.title('Взаимосвязь цены и жилой площади после ремонта');
```
