## Задание 5.
### 5.1 Выберите ноутбук клиенту
Клиент хочет подобрать ноутбук с 8GB или 16GB ОЗУ на Windows 10 в
стоимости до 500 евро, сколько у него вариантов?

```python
df[
    (df['TypeName'] == 'Notebook') &
    (df['Ram'].isin(['8GB', '16GB'])) &
    (df['OpSys'] == 'Windows 10') &
    (df['Price_euros'] <= 500)
]
```

```python
df[
    (df['TypeName'] == 'Notebook') &
    (df['Ram'].isin(['8GB', '16GB'])) &
    (df['OpSys'] == 'Windows 10') &
    (df['Price_euros'] <= 500)
].count()
```
